**Loading the Dataset from the CSV file**

```python
import pandas as pd

tornado = pd.read_csv("Tornado_Data_Count.csv", index_col=0)
tornado.head()
```

**Plotting the Histogram**

**Data Cleansing**

```python
tornado.head()
```

*By analyzing the above dataframe, we could able to see there are many zero values as part of the dataset.*

*Even though there are no occurance of the tornado for a particular, for the purpose of analyzing we are replacing the value of 0's to the  mean values for the respective states.*

```python
for col in tornado:
    tornado[col] = tornado[col].replace(0, round(tornado[col].mean(),2))
```

```python
tornado.head()
```

*Now, we can able to see that the 0's has been replaced with the mean values*

**Methods of Moments**

*Calculating the mean values for all the US states:*

```python
tornado_mean = tornado.mean()
tornado_mean
```

*Calculating the variance for all the US states:*

```python
tornado_var = tornado.var()
tornado_var
```

*Using the moments to find the $\alpha$ and $\beta$ for each month:*

**Calculating the $\alpha$:**

$\alpha = \frac{\bar{X}^2}{S^2}$


Mean: $\bar{X}$ 

Variance: $S^2$ 

```python
alpha_mom = (tornado_mean)**2/tornado_var
alpha_mom
```

**Calculating the $\beta$:**

$\beta = \frac{S^2}{\bar{X}}$


Mean: $\bar{X}$ 

Variance: $S^2$ 

```python
beta_mom = tornado_var/tornado_mean
beta_mom
```

**Plotting the histogram with Mean and Variance for the respective state**

Using the **gamma.pdf** function in **scipy.stats.distributions** to plot the distributions implied by the calculated alphas and betas. 


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.stats.distributions import gamma

tornado['GEORGIA'].hist(bins=20)  # Checking for the state Georgia
plt.plot(np.linspace(0, 10), gamma.pdf(np.linspace(0, 10), alpha_mom[0], beta_mom[0]))
```

Plotting the Histogram for all the states:

*We have increased the size of the histogram (20,10) for better view*

```python
tornado_states = tornado.hist(figsize=(12, 8), sharex=True, sharey=True, bins=15, grid=True)
tornado_states
```

Now, adding the **label values** for all the plotted histogram along with their **mean and variance**

```python
print("US States:\n")
for x in tornado_states.ravel():
    print(x.get_title())
```

```python
tornado_states = tornado.hist(figsize=(20, 15), bins=15, grid=True)

for i in tornado_states.ravel():
    
    m = i.get_title()  #Month
    x = np.linspace(*i.get_xlim())  # Plotting fitted distribution
    i.plot(x, gamma.pdf(x, alpha_mom[m], beta_mom[m]))
    
    label = f'alpha = {round(alpha_mom[m],2)}\nbeta = {round(beta_mom[m],2)}'
    i.annotate(label, xy=(10, 3))
    
plt.tight_layout()
```
